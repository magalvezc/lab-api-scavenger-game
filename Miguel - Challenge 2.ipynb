```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
git3 = requests.get('https://api.github.com/repos/ironhack-datalabs/mad-oct-2018/commits')
git3 = git3.json()
```

```python
git3
```

```python
df = pd.DataFrame(git3)
df
```

```python
df['commit'].count()
```

```python
df_columns = []
for x in df:
    for y in x["commit"]:
        df_columns.append(y)
df = pd.DataFrame(df_columns)
df = json_normalize(df_columns)
df

#buscar dentro de igual que con ejercicio de Ricky
```
